Supervised Learning
Definition:
Supervised Learning is a type of Machine Learning where the data has labels (correct answers). The model learns by comparing its predictions to the known answers and adjusting itself to reduce errors.

Input → Known Output

Example: Email → Spam or Not Spam

Goal: Learn a mapping from input to output.

There are two main types:

Classification (predict categories)

Regression (predict numbers)

Unsupervised Learning 

Definition:
Unsupervised Learning is used when the data does not have labels. The model tries to find hidden patterns, groupings, or structures in the data on its own.

Input → No known output

Goal: Discover structure in data

Common task: Clustering

Reinforcement Learning

Definition:
Reinforcement Learning is a type of learning where an agent learns by interacting with an environment. It receives rewards or penalties based on its actions and improves its strategy over time.

Action → Reward/Penalty → Learn

Goal: Maximize total reward over time.

| Learning Type              | Data Has Labels? | Goal                   | Real-World Example 1  | Real-World Example 2             |
| -------------------------- | ---------------- | ---------------------- | --------------------- | -------------------------------- |
| **Supervised Learning**    | ✅ Yes            | Predict known outcomes | Spam Detection        | House Price Prediction           |
| **Unsupervised Learning**  | ❌ No             | Find hidden patterns   | Customer Segmentation | Market Basket Analysis           |
| **Reinforcement Learning** | Reward-based     | Maximize reward        | Game Playing AI       | Self-Driving Car Decision Making |
