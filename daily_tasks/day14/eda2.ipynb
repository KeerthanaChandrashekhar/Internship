```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde  #library
```

```python
tips = sns.load_dataset('tips')
#loading the dataset
```

```python
tips.describe()
#to understand the data
```

```python
sns.displot(tips.total_bill, kde =True)  #histogram with density line
```

```python
sns.jointplot(x=tips.tip, y=tips.total_bill)   #joining 2 plots to understand data
```

```python
sns.jointplot(x= tips.tip, y= tips.total_bill, kind ='reg')  #tip and total bill understand
```

```python
sns.jointplot(x= tips.tip, y= tips.total_bill, kind ='hex')
```

```python
sns.pairplot(tips, kind= 'reg')
```

```python
sns.pairplot(tips, hue='time')
```

```python
sns.pairplot(tips, hue='day')
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True)
```

```python
sns.boxplot(tips.total_bill)
sns.boxplot(tips.tip)
```

```python
sns.countplot(x=tips.day)
sns.countplot(x=tips.sex)
```

```python
tips.sex.value_counts().plot(kind='pie', autopct= '%1.1f%%')
tips.sex.value_counts().plot(kind='bar')
```

```python
sns.countplot(data=tips[tips.time == 'Dinner'], x='day')
sns.countplot(data=tips[tips.time == 'Lunch'], x='day')
```

```python
fg = sns.FacetGrid(tips, row= 'smoker', col='time')
fg.map(sns.histplot, 'total_bill')
```
