```python
import numpy as np 
import matplotlib.pyplot as plt 

data = np.array([10,12,13,14,15, 16,17,18,19,100,105])
```

```python
data
```

```python
mean = np.mean(data)

std_val = np.std(data)
```

```python
fig = plt.figure()

plt.boxplot(data)
plt.show()

```

```python
z_scores = (data - mean) / std_val
print("Z-scores:", z_scores)
print("Mean:", mean)
print("Standard Deviation:", std_val)
```

```python
plt.figure()
plt.hist(data)
plt.title("Histogram of Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
